## Cálculo da Anomalia Bouguer para o mundo todo

```python
fname = 'dataset/topography_etopo1.gdf'
```

```python
data = hm.load_icgem_gdf(fname)
```

```python
topography = data.topography_shm
```

```python
fname = 'dataset/geoid_XGM2019e.gdf'
```

```python
data = hm.load_icgem_gdf(fname)
```

```python
geoid = data.geoid
```

```python
height = geoid + topography
```

```python
bouguer = hm.bouguer_correction(height)
anomaly = disturbance - bouguer
```

### Anomalia bouguer para o mundo inteiro

```python
fig = pygmt.Figure()
fig.grdimage(
    anomaly,
    projection="W20c",
    cmap="polar+h",
    shading="+a45+nt0.2",
)
fig.basemap(frame=["af", "WEsn"])
fig.colorbar(
    position="JCB+w10c",
    frame=["af", 'y+l"mGal"', 'x+l"Bouguer anomaly"'],
)
fig.coast(shorelines=True, resolution="c", area_thresh=1e4)
fig.savefig('images/bouguer_anomaly.png', dpi=300)
fig.show()
```
