# 13ª Geração do Campo Geomagnético de Referência (IGRF-13)

## Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pygmt 
```

## Carregando os dados de inclinação, declinação, componente horizontal e campo total

```python
fname_inc = 'dataset/igrf_grid_inclination.csv'
names = ['year', 'latitude','longitude','inclination']
```

```python
inc = data = pd.read_csv(fname_inc,usecols=(0,1,2,4),names=names,skipinitialspace=True,skiprows=13)
```

```python
inc
```

## Armazenando dados em variáveis

```python
lat = inc.latitude.values
long = inc.longitude.values
inclination = inc.inclination.values
```

```python
inclination
```

## Configuração do campo para o ano de 2020 

```python
fig = pygmt.Figure()
fig.basemap(region="d", projection="W25c", frame=["af", "WEsn"])
fig.coast(shorelines=True, resolution="c", area_thresh=1e4,land="gray", water="lightblue")
fig.contour(
    projection="W25c",
    frame=["ag","lSEt+tInclination (IGRF-13)"],
    pen="0.5p",
    x = long,
    y = lat,
    z = inclination,
    levels=10,
    annotation=20,
)
fig.savefig('images/inclination_igrf_13.png', dpi=200)
fig.show()
```
